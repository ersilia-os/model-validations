```python
import os
import pandas as pd
from rdkit import Chem
import matplotlib.pyplot as plt

DATAPATH = "../data"
SMICOL = "smiles"
INCHICOL = "inchikey"
ACTCOL = "activity"
```

## Model Bias
### This notebook checks for the model bias of the predictions of the models against the validations csv file [here](https://github.com/Richiio/model-validations/blob/main/data/Validations.csv)

## eos2ta5

```python
# Read the CSV file into a pandas DataFrame
predictions_df = pd.read_csv(os.path.join(DATAPATH,"bias_predictions", "bias_predictions_eos2ta5.csv"))  # Replace 'predictions.csv' with your file path

prediction_column = 'probability'

# Plotting the histogram
plt.hist(predictions_df[prediction_column], bins=30, color='skyblue', edgecolor='black')

# Adding labels and title
plt.xlabel('Predicted Values')
plt.ylabel('Frequency')
plt.title('Distribution of Predicted Values')

# Showing plot
plt.show()
```
